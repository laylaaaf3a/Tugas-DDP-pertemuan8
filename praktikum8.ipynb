```python
#nomor 2
def kelulusan(nilai) :
  if nilai < 60 :
    return "gagal"
  elif 60 <= nilai <= 70 :
    return "baik"
  elif 71 <= nilai <= 80 :
    return 'sangat baik'
  elif 81 <= nilai <= 100 :
    return 'istimewa'

nilai_siswa = float(input('Masukan nilai Mahasiswa:'))
hasil = kelulusan(nilai_siswa)
print('hasil Nilai =', hasil)
```

```python
#nomor 3
def ganjil(n) :
  for i in range(1, n+1, 2) :
    print(i)

ganjil(100)
```
